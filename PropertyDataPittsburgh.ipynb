```python
# load pandas
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
property_data = pd.read_csv("https://data.wprdc.org/datastore/dump/8eff881d-4d28-4064-83f1-30cc991cfec7")
```

```python
property_data
```

This dataset has a lot of columns of data that are not all that meaningful to what I am trying to go for. To start, I will filter the dataset into a cleaner dataset with the columns that I will need.

```python
column_names = ["PROPERTYADDRESS", "NEIGHBORHOOD", "SALEPRICE", "BEDROOMS"]
house_data = pd.DataFrame(property_data, columns=column_names)
```

```python
house_data.head(5)
```

Now that I have the data into an easier to look at set, I want to filter out certain entries that are not important as well. I am looking for a neighborhood, so I will filter out the rows that don't have a listed neighborhood. This data set also has the overall property sales which could be empty lots, buildings, stores and things of that nature. I want to get just properties to live in so I will filter based on number of bedrooms.

```python
houses = house_data[house_data['BEDROOMS'].notna()]
```

```python
houses = houses[houses['NEIGHBORHOOD'].notna()]
```

```python
houses
```

Now that I have a cleaned up dataset I first want to find the overall average. Next I will group each neighborhood and find their individual average.

```python
total_sum = houses["SALEPRICE"].sum()
```

```python
avg_n = dict(houses.groupby('NEIGHBORHOOD')['SALEPRICE'].mean())
```

```python
avg_n
```

```python
avg_info = {'Average': avg_n}
```

```python
avgF = pd.DataFrame(avg_info)
```

Here I take the data and put it into a data frame to be able to sort it. 

```python
avgF = avgF.sort_values(by=['Average'], ascending=False)
avgF
```

```python
graph = avgF.plot.barh(figsize=(15,avgF.shape[0] * .5), legend=None)
```

```python
avgF['Score'] = (avgF['Average']/10000)*3
avgF
```
